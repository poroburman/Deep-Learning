```python
# This script uses ANN to analyze and predicts which customers continue to keep their accounts within a specific bank
# and which customers leave the bank
# The dataset contains the details of all the customers within a bank
```

```python
# 1. DATA PREPROCESSING
```

```python
# Import libraries
import os
import numpy as np
import pandas as pd
import tensorflow as tf

# set the working directory
os.chdir('/Users/poromendro/Desktop')

# import the dataset
dataset = pd.read_csv('Churn_Modelling.csv')

# set the independent and dependent variables
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values

# encode categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))

# split dataset into training and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# feature scaling 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# 2. BUILD THE ANN
```

```python
# create ann instance
ann = tf.keras.models.Sequential()

# add input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))

# add second hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))

# add the output layer
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

```python
# 3. TRAINING THE ANN
```

```python
# compiling the ANN
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ["accuracy"])

# training the ANN on the training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
# 4. MAKE A PREDICTION
```

```python
# make a prediction whether the customer with below credentials leaves the bank
# Geography: France
# Credit Score: 600
# Gender: Male
# Age: 40 years old
# Tenure: 3 years
# Balance: $ 60000
# Number of Products: 2
# Does this customer have a credit card ? Yes
# Is this customer an Active Member: Yes
# Estimated Salary: $ 50000
print(f"1. Does the customer leave the bank? {ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]])) > 0.5}")
print("\n")

# Predict test set results
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(f"2. np array with predicted values in the first column and test values in the second column \n\n {np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)}")
print("\n")

# Make the confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(f"3. confusion matrix: \n {cm}")
print("\n")
print(f"4. The accuracy score for this model is : {accuracy_score(y_test, y_pred)}")
```
